```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import ast
import csv
```

```python
train_english = '../Datasets/train2.0_english.csv'

df = pd.read_csv(train_english)
```

```python
genre_dict = {}
for i, d in df.iterrows():
    genre_list = ast.literal_eval(d.genre)
    for genre in genre_list:
        genre_name = genre['genre_name']
        if genre_name in genre_dict:
            genre_dict[genre_name] = genre_dict[genre_name] + 1
        else:
            genre_dict[genre_name] = 1
            
genre_dict
```

```python
top_amount = 20
top_genres = sorted(genre_dict.items(), key=lambda x: x[1], reverse=True)[:top_amount]

print(top_genres)

plt.bar(range(top_amount), [gen[1] for gen in top_genres])
plt.title(f"Top {top_amount} genres distribution")
plt.xticks(range(top_amount), [gen[0] for gen in top_genres], rotation=60) 
plt.ylabel('Count')
plt.show()
```
